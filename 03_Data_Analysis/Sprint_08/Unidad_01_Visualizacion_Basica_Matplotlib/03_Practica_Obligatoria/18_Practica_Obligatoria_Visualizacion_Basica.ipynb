1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center" o investigar como funciona bar_label). Los colores de las barras deben tener un 50% de transparencia y ser: azul, gris, rojo, aceituna y marrón. 

```python
top5 = df_cities.nlargest(5, "population_total")[["city", "population_total"]]
colores = ["blue", "gray", "red", "olive", "brown"]

fig1, ax1 = plt.subplots(figsize=(6, 5)) 

bars = ax1.bar(
    top5["city"],
    top5["population_total"] / 1000, # Dividimos entre 1000 si tus datos están en unidades
    color=colores,
    alpha=0.5
)

for bar in bars:
    height = bar.get_height()
    ax1.text(
        bar.get_x() + bar.get_width() / 2, 
        height + 20,                       
        f'{height:.2f}',                   
        ha="center",                       
        va="bottom",                       
        fontsize=9,
    )


ax1.set_title("Ciudades más pobladas de California")
ax1.set_xlabel("Ciudades")
ax1.set_ylabel("Habitantes (miles)")
ax1.set_ylim(0, 5000);

```

2. Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Arial", tamaño = 14, peso = "Bold", color azul acero o "steelblue")

```python
fig2, ax2 = plt.subplots(figsize=(5, 4))

ax2.hist(df_cities["area_total_sq_mi"], 
         bins = 50, 
         color = "steelblue", 
         edgecolor = "white"
)

ax2.set_title("Histograma del Área Total de las Ciudades",
              fontname = "Arial", 
              fontsize = 14, 
              fontweight = "bold"
)

ax2.set_xlabel("Área total ($km^2$)", 
               fontname = "Arial", 
               fontsize = 14, 
               fontweight = "bold"
)

ax2.set_ylabel("Frecuencia", 
               fontname = "Arial", 
               fontsize = 14, 
               fontweight = "bold"
)

plt.tight_layout()
```

3. El scatter de área de agua de un ciudad con su poblacion. Transparencia al 50% y color verde.

```python
plt.figure(figsize = (4, 4))

plt.scatter(
    df_cities["area_water_km2"], 
    df_cities["population_total"] / 1000, 
    color = "green", 
    alpha = 0.5
)

outliers = df_cities.nlargest(2, "population_total")

for i, row in outliers.iterrows():
    plt.annotate(
        "Outliers", 
        xy = (row["area_water_km2"], row["population_total"] / 1000),
        xytext = (65, 3000), 
        arrowprops = dict(facecolor = "black", arrowstyle = "->"),
        ha = "center"
    )

plt.title("Población por kilometro cuadrado de agua")
plt.xlabel("Área de agua en la ciudad ($km^2$)")
plt.ylabel("Población (miles habitantes)")
plt.ylim(-200, 4200) 
```

4. El boxplot de la altura de las ciudades.

```python
plt.figure(figsize = (6, 6))

plt.boxplot(df_cities["elevation_m"], 
            vert = False, 
            patch_artist = True,
            boxprops = dict(facecolor = "red", color = "black"), # Caja roja con borde negro
            medianprops = dict(color = "black")
)

plt.title("Diagrama de caja de la Altitud de las ciudades de California")
plt.xlabel("Altitud (m)")

# Quitamos los números del eje Y porque en la imagen no aparecen
plt.yticks([])
plt.xlim(-200, 2500)
```

### Figuras combinadas

```python
fig, ax = plt.subplots(2, 2, figsize=(16, 14))

# Figura 1
top5 = df_cities.nlargest(5, "population_total")
colores = ["blue", "gray", "red", "olive", "brown"]
bars = ax[0,0].bar(top5["city"], top5["population_total"] / 1000, color=colores, alpha=0.5)

for bar in bars:
    height = bar.get_height()
    ax[0,0].text(bar.get_x() + bar.get_width() / 2, height + 20, f'{height:.2f}', 
                 ha="center", va="bottom", fontsize=9)

ax[0,0].set_title("Ciudades más pobladas de California")
ax[0,0].set_xlabel("Ciudades")
ax[0,0].set_ylabel("Habitantes (miles)")
ax[0,0].set_ylim(0, 5000)


# Figura 2
ax[0,1].hist(df_cities["area_total_km2"], bins=50, color="steelblue", edgecolor="white")

font_style = {'fontname': 'Arial', 'fontsize': 14, 'fontweight': 'bold'}
ax[0,1].set_title("Histograma del Área Total", **font_style)
ax[0,1].set_xlabel("Área total ($km^2$)", **font_style)
ax[0,1].set_ylabel("Frecuencia", **font_style)


# Figura 3
ax[1,0].scatter(df_cities["area_water_km2"], df_cities["population_total"] / 1000, 
                color="green", alpha=0.5)

outliers = df_cities.nlargest(2, "population_total")
for i, row in outliers.iterrows():
    ax[1,0].annotate("Outliers", 
                     xy=(row["area_water_km2"], row["population_total"] / 1000),
                     xytext=(65, 3000), 
                     arrowprops=dict(facecolor="black", arrowstyle="->"),
                     ha="center")

ax[1,0].set_title("Población por kilometro cuadrado de agua")
ax[1,0].set_xlabel("Área de agua en la ciudad ($km^2$)")
ax[1,0].set_ylabel("Población (miles habitantes)")
ax[1,0].set_ylim(-200, 4200)


# Figura 4
ax[1,1].boxplot(df_cities["elevation_m"], vert=False, patch_artist=True,
                boxprops=dict(facecolor="red", color="black"),
                medianprops=dict(color="black"))

ax[1,1].set_title("Diagrama de caja de la Altitud")
ax[1,1].set_xlabel("Altitud (m)")
ax[1,1].set_yticks([])
ax[1,1].set_xlim(-200, 2500)


plt.tight_layout() # para que no se solapen

```
